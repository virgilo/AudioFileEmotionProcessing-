```python
###J'IMPORTE MON MODULE QUE J AI CREER POUR LE AUDIO PROCESSING
from  Mon_Module_Audio import *
##Conversations Paths
ListOfConversations=[]
Conversations_Path="./ArtificialdatasetAudiosConversation/zConversationsUsed"
ListOfConversations=ListOfWavFileInaFolder(Conversations_Path) 
ListOfConversations=SortAlphanumericalList(ListOfConversations) ##JE TRIE ListOfConversations selon l'ordre croissant du suffixe numérique _OOO

#le module pour  segmenté les conversations en se basant sur le SILENCE

# Import the AudioSegment class for processing audio and the 
# split_on_silence function for separating out silent chunks.
from pydub import AudioSegment
from pydub.silence import split_on_silence
from  python_speech_features import mfcc
print('DEBBBBBBBbb')
 
for k,uneConversation_Path in enumerate(ListOfConversations):
    # Load your audio.
    song = AudioSegment.from_wav(uneConversation_Path)

    ## Conversion AudioSeg en Array
    song_ARRAY=AudioSegToArray(song)
    #############################################song_ARRAY=AudioSegToNumpyArray(song)

    ###Fct pour Segmenter TAF
    #Remarquer on va segmenter en fct des du separateur Marker_Array
    ##Obtenir les intervalle d'index à extraire
    List_Index_Chunks=list()

    Index_Start=0
    Index_End=0
    i=0
    print('bEGINNG') 
    while (Index_Start+len(Marker_Array) < len(song_ARRAY)):
        print('XX')
        while ((i+len(Marker_Array)<=len(song_ARRAY)) and (song_ARRAY[i:i+len(Marker_Array)] != Marker_Array)):
            #print('YY')
            i=i+1 
        Index_End=i
        if (i+len(Marker_Array) >len(song_ARRAY)):
            Index_End=len(song_ARRAY)
        Indexs_chunk=[Index_Start , Index_End]
        List_Index_Chunks.append(Indexs_chunk)    
        Index_Start=i+len(Marker_Array)
        i=i+len(Marker_Array)
        print(len(List_Index_Chunks))

    ##Obtention des segments d'audios correspondants aux intervalles de prises de paroles obtenues
    i=0
    indiceRepertoirePriseDeParole=1
    for  Index_Chunks in List_Index_Chunks:
        PriseDeParole_Array=song_ARRAY[Index_Chunks[0]:Index_Chunks[1]]
        PriseDeParole_NUMPY_ARRAY=np.int16(PriseDeParole_Array)
        PriseDeParole_AudioSeg=NumpyArrayToAudioSeg(PriseDeParole_NUMPY_ARRAY)
        ##EXPORTATION EN FICHIER WAV
        # Export the audio chunk with new bitrate.
        if (i%2==0): #SpeakerO        
            print("bExtractedCreatedSegmentSpeech1O{0}.wav.".format(i))
            PriseDeParole_AudioSeg.export(
                "./AudiosApresSegmentation/AudiosDuLocuteurs1/Conv_"+str(k)+"_Loc_1_ExtractedUtt_"+str(i)+".wav",
                bitrate = "192k",
                format = "wav"
            )
            #Stocker une copie dans le repertoire de comparaison des segment pour validation
            PriseDeParole_AudioSeg.export(
                "./Fichiers de comparaison test/LOCUTEUR O/Par nature-Initiaux Vs Segments/Segments/"+str(indiceRepertoirePriseDeParole)+"/bLocuteur1OExtractedElementaryAudios{0}.wav".format(i),
                bitrate = "192k",
                format = "wav"
            )
        else:
            print("bExtractedCreatedSegmentSpeech2Y{0}.wav.".format(i))
            PriseDeParole_AudioSeg.export(
                "./AudiosApresSegmentation/AudiosDuLocuteurs2/Conv_"+str(k)+"_Loc_2_ExtractedUtt_"+str(i)+".wav",
                bitrate = "192k",
                format = "wav"
            )
            #Stocker une copie dans le repertoire de comparaison des segment pour validation
            PriseDeParole_AudioSeg.export(
                "./Fichiers de comparaison test/LOCUTEUR Y/Par nature-Initiaux Vs Segments/Segments/"+str(indiceRepertoirePriseDeParole)+"/bLocuteur2YExtractedElementaryAudios{0}.wav".format(i),
                bitrate = "192k",
                format = "wav"
            )
            indiceRepertoirePriseDeParole=indiceRepertoirePriseDeParole+1
        i=i+1

    
print('SEGMENTATION FINISHED ALL')
```
